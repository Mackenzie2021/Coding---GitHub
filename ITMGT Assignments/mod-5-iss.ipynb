```python
import requests
import urllib.request
from urllib.request import urlopen
import pandas as pd
import numpy as np
import json
import time
import matplotlib.pyplot as plt
```

```python
lat_list=[]
long_list=[]
time_list=[]
```

```python
for i in range(0,121):
    link_tracker = "http://api.open-notify.org/iss-now.json"
    response = urllib.request.urlopen(link_tracker)
    result = json.loads(response.read())

    iss_position = result["iss_position"]

    longitude = float(iss_position['longitude'])
    long_list.append(longitude)

    latitude = float(iss_position['latitude'])
    lat_list.append(latitude)

    timemark = result['timestamp']
    time_list.append(timemark)
    
    time.sleep(10)
    
d = {'Latitude': lat_list, 'Longitude': long_list, 'Timestamp': time_list}
df = pd.DataFrame( data =d )
df
```

```python
fig = plt.figure(figsize=(50,50))
ax = fig.add_subplot()

plt.title("ISS Position")
ax.scatter(lat_list,long_list,color="red")
ax.set_xlim([-100, 100])
ax.set_xlim([-100, 100])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
```
